```python
#Exercício - 03
meses = int(input("Digite quantos meses terá a aplicação: "))
valor = int(input("Digite o valor da aplicação: "))
taxa = 1.1
cont = 1
while (cont <= meses):
 print(f'{cont}° mês: {valor} * {taxa} = {round(valor*taxa,2)}')
 valor = round(valor*taxa,2)
 cont += 1
```

```python
#Exercício - 05
meses = int(input("Digite quantos meses terá a aplicação: "))
valor = int(input("Digite o valor da aplicação: "))
part = int(input("Digite a quantidade de participantes: ")) 
taxa = 1.02
cont = 1
print(cont,"° mês: ",valor,"*",part,"=",valor*part)
cont = 2
while (cont <= meses):
 print(cont,"° mês: (",valor,"*",taxa,") *",part,"=",round((valor*taxa)*part,2))
 valor = round(valor*taxa,2)
 cont += 1
```
