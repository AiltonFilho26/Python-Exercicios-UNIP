```python
#Exercício C

def calc_idade(x, ano):
  return ano - x

def m_18(x):
  if x <= 18:
    return True

def m_30(x):
  if 18 < x <= 30:
    return True

def m_50(x):
  if 30 < x <= 50:
    return True

def m_80(x):
  if 50 < x <= 80:
    return True

ano = 2022
x = 1
while x != 0:
  x = int(input('Digite o ano de nascimento da próxima pessoa que será cadastrada (digite 0 para finalizar): '))
  calc_idade(x,ano)
  if m_18(calc_idade(x,ano)):
    print('É uma pessoa menor de idade, possui {} anos.'.format(calc_idade(x,ano)))
  elif m_30(calc_idade(x,ano)):
    print('É uma pessoa maior de idade com menos de 30 anos, possui {} anos.'.format(calc_idade(x,ano)))
  elif m_50(calc_idade(x,ano)):
    print('É uma pessoa com idade entre 30 e 50 anos, possui {} anos.'.format(calc_idade(x,ano)))
  elif m_80(calc_idade(x,ano)):
    print('É uma pessoa com idade entre 50 e 80 anos, possui {} anos.'.format(calc_idade(x,ano)))
  elif x == 0:
    print('Finalizando cadastro!')
    break
```

```python
#Exercício D
def calculo():
  soma = 0
  for x in range (1,11):
    if x % 2 == 0:
      x = -x
    soma += x
    print(x)
  return soma
print('O resultado da operação entre os números de 1 a 10 sendo os pares negativos é: {}'.format(calculo()))
```

```python
#Exercício E
def calculo():
  soma = 0
  divi = 0
  x = 1
  for y in range (10,0,-1):
    if y % 2 == 1:
      y = -y
    divi = x/y
    soma = soma + divi
    print('{}/{} = {}'.format(x,y,round(divi,3)))
  print('Resultado final = {}'.format(round(soma,3)))
calculo()
```
