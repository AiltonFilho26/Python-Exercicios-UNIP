```python
#Exercício B
d = int(input('Digite a quantidade de dias de viagem: '))
def orcamento_1(d):
  return 220 + 129*d
def orcamento_2(d):
  return 173 + 162*d
print('Com o primeiro orçamento você irá pagar: R${}.'.format(orcamento_1(d)))
print('Com o segundo orçamento você irá pagar: R${}.'.format(orcamento_2(d)))
print('Com o primeiro orçamento sairá R${} mais barato.'.format(orcamento_2(d)-orcamento_1(d)))
```

```python
#Exercício C
def juros(valor, tx_juro):
  return valor * tx_juro
valor = 134.0
tx_juro = 1.03
mes = 0
print('No 1º mês o valor será: R${}'.format(juros(valor, tx_juro)))
while mes < 3:
  valor = juros(valor, tx_juro)
  mes = mes + 1
print('No {}º mês o valor será: R${}.'.format(mes,round(valor,2)))
while valor <= 160:
    valor = juros(valor, tx_juro)
    mes = mes + 1
print('No {}º mês o valor será: R${}.'.format(mes,round(valor,2)))
while valor <= (134*2):
  valor = juros(valor, tx_juro)
  mes = mes + 1
print('No {}º mês o valor será: R${}, o dobro do valor inicial.'.format(mes,round(valor,2)))
```

```python
#Exercício E
print('-'*87)
print('Sua senha deve conter de 6 a 12 letras, um número, uma letra maiúscula e uma minúscula!')
print('-'*87)


while True:
  senha = input('Digite sua senha: ')
  print('-'*50)
  senha2 = input('Confirme sua senha: ')
  print('-'*50)
  if senha2 == senha:
    if len(senha) < 6:
      print('Deve conter no mínimo 6 caracteres.')
    elif len(senha) > 12:
      print('Deve conter no máximo 10 caracteres.')
    else:
      if not any(x.isupper() for x in senha):
        print('Deve conter pelo menos uma letra maiúscula.')
      elif not any(x.islower() for x in senha):
        print('Deve conter pelo menos uma letra minúscula.')
      elif not any(x.isdigit() for x in senha):
        print('Deve conter pelo menos um número.')
      else:
        break
  else:
    print('Senhas diferentes.')
    print('Tente novamente.')
    print('-'*50)

print('Senha CADASTRADA!')
```
