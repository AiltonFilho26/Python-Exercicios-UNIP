```python
#Exercício 01 - c
valor = float(input('Coloque aqui o valor da sua compra: '))
if valor > 150.00:
  desconto = valor*0.1

elif valor == 150.00:
  desconto = valor*0.07

elif valor < 150.00:
  desconto = valor*0.04
  
print('O valor com desconto é: {:.2f}'.format(valor-desconto))
```
