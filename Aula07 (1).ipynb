```python
#Exercício 01 - d
#Crie uma senha com somente letras maiusculas ou somente letras minusculas.
while True:
  senha = str(input('Digite a senha: '))
  if senha.islower():
    print('Senha válida.')
    break
  elif senha.isupper():
    print('Senha válida.')
    break
  else:
    print('Senha inválida, tente novamente.')
```

```python
#Exercício 01 - e
#Escreva uma string e coloque os caracteres de posição par maiusculos 
#e os caracteres de posição ímpar minusculos.
frase = str(input('Digite uma frase: '))  
vazio = ''
for posicao in range(len(frase)):
  if posicao % 2 == 0:
    vazio += frase[posicao].upper()
  else: 
    vazio += frase[posicao].lower()
print(vazio)
```

```python
#Exercício 01 - f
#Manipule uma string de forma que inverta os caracteres maiusculos por 
#minusculos e vice-versa sem utilizar o comando 'swapcase'.
frase = str(input('Digite uma frase: '))
vazio = ''
for c in frase:
  if c.isupper():
    vazio += c.lower()
  else:
    vazio += c.upper()

print(vazio)
```

```python
#Exercício 01 - g
#Conte quantas vogais e quantas consoantes tem em uma string qualquer.
vogais = ('a','e','i','o','u')
texto = input('Digite uma frase: ').lower().replace(" ","")
cont_vogais = 0
cont_consoantes = 0
for letra in texto:
  if letra in vogais:
    cont_vogais += 1
  else:
    cont_consoantes += 1
print('Vogais: %d' % cont_vogais)
print('Consoantes: %d' % cont_consoantes)
```

```python
#Exercício 01 - h
#Faça um código em que o usuário coloque uma cadeia de DNA para o programa
#gerar a cadeia complementar.
dna = str(input('Digite a cadeia de DNA: ')).upper()
cc = ''
funcional = ('T','A','C','G')
for letra in dna:
  if letra not in funcional:
    print('Cadeia inexistente')
    break
  else:
    if letra == 'T':
      cc += 'A'
    elif letra == 'A':
      cc += 'T'
    elif letra == 'C':
      cc += 'G'
    elif letra == 'G':
      cc += 'C'
print('Saída: %s' %cc)
```

```python
#Exercício 01 - i
#Escreva uma função que recebe uma frase, uma palavra antiga e uma palavra nova. A função deve retornar uma
#string contendo a frase original, mas com a última ocorrência da palavra antiga substituída pela palavra nova. A
#entrada e saída de dados deve ser feita no programa principal. 
frase = input('Digite uma frase: ').lower()
while True:
  pa = input('Qual a palavra a ser trocada? ').lower()
  if pa in frase:
    pn = input('Qual a palavra nova? ').lower()
    print(frase.replace(pa,pn).capitalize()) 
    break
  else: 
    print('Essa palavra não existe na frase.')
```

```python
#Exercício 01 - j
#Faça um código que recebe uma frase e substitui todas as ocorrências de espaço por “#”.
frase = str(input('Digite uma frase: '))
print('#'.join(frase.split(' ')))
```

```python
#Exercício 03
matriz = []
for x in range(4):
  linha = []
  for y in range(4):
    linha.append(int(input('Digite o valor da posição [{},{}]:'.format(str(x),str(y)))))
  matriz.append(linha)
impares = 0
for x in range(4):
  for y in range(4):
    if matriz[x][y] % 2 == 1:
      impares += 1

for x in range(4):
  print(matriz[x])
print('Nessa matriz possui %d números ímpares.'%impares)
```
